```python
import pandas as pd 
import numpy as np
train=pd.read_csv("/content/laptops_train.csv")
test=pd.read_csv("/content/laptops_test.csv")
train
```

```python
train.isnull().sum()
```

```python
train= train.drop(['Operating System Version','Model Name','Category','Weight'], axis=1)
train
```

```python
train['CPU core'] = train['CPU'].apply(lambda x: x.split()[0])
train['CPU2'] = train['CPU'].apply(lambda x: x.split()[-1])
train['GHZ'] = train['CPU2'].apply(lambda x: x.split('G')[0])
train = train.drop('CPU2', axis=1)
train = train.drop('CPU', axis=1)
train['Screen Size1'] = train['Screen Size'].apply(lambda x: x.split('\"')[0])
train = train.drop('Screen Size', axis=1)
train['Screen1'] =train['Screen'].apply(lambda x: x.split()[-1])

train['Screen11'] = train['Screen1'].apply(lambda x: x.split('x')[0])
train['Screen12'] = train['Screen1'].apply(lambda x: x.split('x')[1])
train = train.drop('Screen', axis=1)
train = train.drop('Screen1', axis=1)
train['ram']=train['RAM'].apply(lambda x:x.split('G')[0])
train=train.drop("RAM", axis=1)
train
```

```python
test.isnull().sum()
```

```python
test= test.drop(['Operating System Version','Model Name','Category','Weight'], axis=1)
test
```

```python
test['CPU core'] = test['CPU'].apply(lambda x: x.split()[0])
test['CPU2'] = test['CPU'].apply(lambda x: x.split()[-1])
test['GHZ'] = test['CPU2'].apply(lambda x: x.split('G')[0])
test = test.drop('CPU2', axis=1)
test = test.drop('CPU', axis=1)
test['Screen Size1'] = test['Screen Size'].apply(lambda x: x.split('\"')[0])
test = test.drop('Screen Size', axis=1)
test['Screen1'] =test['Screen'].apply(lambda x: x.split()[-1])

test['Screen11'] = test['Screen1'].apply(lambda x: x.split('x')[0])
test['Screen12'] = test['Screen1'].apply(lambda x: x.split('x')[1])
test = test.drop('Screen', axis=1)
test = test.drop('Screen1', axis=1)
test['ram']=test['RAM'].apply(lambda x:x.split('G')[0])
test=test.drop("RAM", axis=1)
test
```

```python
df = pd.concat([train,test])
df
```

```python
train= pd.get_dummies(train, columns = ['Manufacturer',' Storage','GPU','Operating System','CPU core']) 
train
```

```python
test= pd.get_dummies(test, columns = ['Manufacturer',' Storage','GPU','Operating System','CPU core']) 
test
```

```python
X_train = train[train.columns[1:168]]
y_train = train.iloc[:977, 0]   
X_test = test.iloc[:977, 1:]  
y_test = test.iloc[:977, 0] 
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import LabelEncoder 
regressor= LinearRegression()  
regressor.fit(X_train, y_train)  
```

```python
print('Train Score: ', regressor.score(X_train, y_train)) 
from sklearn.model_selection import cross_val_score
print("cross validation output:",cross_val_score(regressor,X_test,y_test, cv=3)) 
```
